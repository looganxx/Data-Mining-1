## Assegno i valori mancanti di time_signature
Ad ogni traccia che ha un determinato genere, il valore di time_signature medio delle canzoni con lo stesso genre.

### Vedo prima come sono distribuiti i valori di time_signature tra i generi: se c'è una distribuzione equa allora posso usare la mediana di ogni genere per riempire i valori vuoti

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="genre", y="time_signature", data=df)
plt.xticks(rotation=45)
plt.title("Distribuzione di time_signature per genere")
plt.show()
```

Da qui si evince che il valore predominante per tutti i generi è il 4.0, posso quindi usare la mediana come valore per i missing

Aggiungo anche la distribuzione generale su tutto il df - non per genere

```python
# Creare il grafico a barre
plt.figure(figsize=(10, 6))
df['time_signature'].value_counts().plot(kind='bar', color='green')
plt.xlabel('Valore time_signature')
plt.ylabel('Frequenza')
plt.title('Distribuzione time_signature')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Sostituisco i valori mancanti di time_signature

```python
# Calcolare la mediana del time_signature per ogni genere
print(df.groupby('genre')['time_signature'].median())
```

```python
time_signature_median = df['time_signature'].median()
print(time_signature_median)
```

```python
df['time_signature'].fillna(time_signature_median, inplace=True)
```

## Voglio calcolare la correlazione tra le feature e genere

### Inizio campionando genere

```python
genre_locks = sorted(df['genre'].unique())
genre_locs_mapping = dict(zip(genre_locks, range(1, len(genre_locks)+1)))
df['genre_val'] = df['genre'].map(genre_locs_mapping).astype(int)
df[['genre', 'genre_val']]
```

```python
df.info()
```

```python
correlation_matrix = df._get_numeric_data().corr()

# Visualizzare la matrice di correlazione 
plt.matshow(correlation_matrix, cmap='coolwarm')
plt.colorbar()
plt.show()

# Analizzare i valori di correlazione individuali
for feature in df._get_numeric_data().columns[:-1]:  # Escludere la colonna "genere"
    correlation_coefficient = correlation_matrix[feature]['genre_val']
    print(f"Correlazione tra {feature} e genere: {correlation_coefficient:.2f}")
```
